```python
import fitz  # PyMuPDF

# 並べたいPDFファイル
pdf_files = [
    "../fig/extract_area_tohoku.pdf", "../fig/fig2_b.pdf",
    "../fig/HK2020_distHRA.pdf", "../fig/090-70_distHRA.pdf",
    "../fig/080-60_distHRA.pdf", "../fig/070-50_distHRA.pdf"
]

# 出力ファイル
output_pdf = "../fig/combined_contour.pdf"

# 用紙サイズ（A4縦）
page_width = 595   # pt
page_height = 842  # pt

# 余白
margin_x = 30
margin_y = 30
gap_x = 10
gap_y = 10

# グリッド
cols = 2
rows = 3

# 各枠サイズ
cell_width = (page_width - 2 * margin_x - gap_x * (cols - 1)) / cols
cell_height = (page_height - 2 * margin_y - gap_y * (rows - 1)) / rows

# 出力PDF作成
out_doc = fitz.open()
page = out_doc.new_page(width=page_width, height=page_height)

for i, pdf_file in enumerate(pdf_files):
    r = i // cols
    c = i % cols

    x0 = margin_x + c * (cell_width + gap_x)
    y0 = margin_y + r * (cell_height + gap_y)
    x1 = x0 + cell_width
    y1 = y0 + cell_height

    rect = fitz.Rect(x0, y0, x1, y1)

    src = fitz.open(pdf_file)
    # 各PDFの1ページ目を配置
    page.show_pdf_page(rect, src, 0)
    src.close()

out_doc.save(output_pdf)
out_doc.close()

print(f"saved: {output_pdf}")
```
